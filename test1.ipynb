```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
# inquiryone.csv is the csv that includes the data from the first experiment
df1 = pd.read_csv("inquiryone.csv")
# defining x and y values (f and v are the centripetal force and velocity squared in the csv file, respectivley)
X = df1['f'].values.reshape(-1, 1)
Y = df1['v'].values.reshape(-1, 1)
# defining regression model from sklearn
reg = LinearRegression()
reg.fit(X, Y)
Y_pred = reg.predict(X)
# r squared from sk metrics
r2 = r2_score(Y, Y_pred)
# putting r squared value into a string to display in legend
equation_str = f"y = {reg.coef_[0][0]:.2f}x + {reg.intercept_[0]:.2f}"
r2_str = f"R² = {r2:.2f}"

plt.scatter(X, Y, color="blue")
plt.plot(X, Y_pred, color="red", label=equation_str)
plt.xlabel("Centripetal Force (N)")
plt.ylabel("Velocity Squared (m/s)")
plt.title("Linear Relationship Between Velocity Squared and Centripetal Force", fontweight='bold', pad=20)
plt.legend()

# r squared value is being placed below the legend
plt.text(X.mean(), Y.max(), r2_str, fontsize=12, ha='center')

plt.show()
```
